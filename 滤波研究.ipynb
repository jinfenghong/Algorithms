```python
import numpy as np
import matplotlib.pyplot as plt

# 生成包含高、中、低频成分的信号
t = np.linspace(0, 1, 1000, endpoint=False)
signal = (5 * np.sin(2 * np.pi * 10 * t) +
          2 * np.sin(2 * np.pi * 50 * t) +
          0.5 * np.sin(2 * np.pi * 200 * t))

# 绘制原始信号
plt.figure(figsize=(10, 4))
plt.title("原始信号")
plt.plot(t, signal)
plt.xlabel("时间")
plt.ylabel("振幅")
plt.grid(True)
plt.show()
```

```python
from scipy.fft import fft, ifft

# 对信号进行傅里叶变换
signal_fft = fft(signal)
```

```python
# 定义一个低通滤波器（例如，保留低频成分）
cutoff_frequency = 30  # 选择一个截止频率
signal_fft[np.abs(signal_fft) > cutoff_frequency] = 0

# 通过逆傅里叶变换获取滤波后的信号
filtered_signal = ifft(signal_fft)

# 绘制滤波后的信号
plt.figure(figsize=(10, 4))
plt.title("滤波后的信号")
plt.plot(t, filtered_signal)
plt.xlabel("时间")
plt.ylabel("振幅")
plt.grid(True)
plt.show()
```
