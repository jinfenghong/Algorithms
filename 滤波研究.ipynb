```python
import numpy as np
import matplotlib.pyplot as plt

# 
t = np.linspace(0, 1, 1000, endpoint=False)
signal = (5 * np.sin(2 * np.pi * 10 * t) +
          2 * np.sin(2 * np.pi * 50 * t) +
          0.5 * np.sin(2 * np.pi * 200 * t))

# 
plt.figure(figsize=(10, 4))
plt.title("")
plt.plot(t, signal)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.show()
```

```python
from scipy.fft import fft, ifft

# 
signal_fft = fft(signal)
```

```python
# 
cutoff_frequency = 30  # 
signal_fft[np.abs(signal_fft) > cutoff_frequency] = 0

# 
filtered_signal = ifft(signal_fft)

# 
plt.figure(figsize=(10, 4))
plt.title("")
plt.plot(t, filtered_signal)
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.show()
```
